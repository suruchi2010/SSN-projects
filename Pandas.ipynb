```python
d1=pd.DataFrame({'key':['A','B'],'val':[1,2]})
d2=pd.DataFrame({'key':['A','B'],'val':[4,5]})
print(d1)
print(d2)
d3=pd.merge(d1,d2,on='key')
print(d3)
```
